# Tổng quan về Logic Mờ

## Định nghĩa hàm mở

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 40, 200)
mu_hot = np.piecewise(x, [x<=25, (x>25)&(x<35), x>=35],
                      [0, lambda x: (x-25)/10, 1])
plt.plot(x, mu_hot)
plt.title("Hàm thuộc của tập mờ 'Nhiệt độ cao'")
plt.xlabel("Nhiệt độ (°C)")
plt.ylabel("Mức độ nóng μ_hot(x)")
plt.grid(True)
plt.show()
```
